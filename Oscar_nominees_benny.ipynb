```python
import scraping_class
import pandas as pd
nominees = 'nominees.txt'
connector = scraping_class.Connector(nominees)
```

```python
from bs4 import BeautifulSoup
import requests

absolute_path = 'https://www.imdb.com/search/title/'
relative_path ='tt6966692/?ref_=adv_li_tt'
url = 'https://www.imdb.com/search/title/?groups=oscar_nominee'
response = connector.get(url, 'nominees_year')

html = response[0].text
soup = BeautifulSoup(html,'html.parser')
```

```python
#Locates and parse the node
table_node = soup.find_all('div', attrs ={'class': 'lister-item-content'})
table_node
```

```python
movie_data = []
for i in table_node:
    try:
        runtime = i.find('span', attrs ={'class':'runtime'}).text
        genre = i.find('span', attrs = {'class':'genre'}).text.strip().split(',')
        metascore = i.find('span', attrs = {'class': 'metascore favorable'}).text
        idx, title, year = i.find('h3', attrs={'lister-item-header'}).text.strip().split('\n')
        link = i.find('a')
        link =str(link).split('href="')[1].split('"')[0]
        link ='https://www.imdb.com' + link
        
        people = i.find_all('p')[2].find_all('a')
        
        actors = []
        
        for actor in people:
            actors.append(actor.text)
        
        try:
            year = year.split('(I) (')[1].split(')')[0]
        except:
            year = year.split('(')[1].split(')')[0]
    except:
        NameError
    movie_data.append([title, link, year, actors[0], actors[1:], runtime, genre, metascore])
```

```python
COLS =['title', 'link', 'year', 'director', 'actors', 'runtime', 'genre', 'metascore']

df = pd.DataFrame(data=movie_data, columns= COLS)
df
```
